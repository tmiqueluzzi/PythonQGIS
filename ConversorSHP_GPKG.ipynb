```python
import geopandas as gpd

shp_path = (
    "/content/drive/MyDrive/"
    "Fpolis - Distritos Adm/"
    "gvw_distritos_administrativos.shp"
)

gpkg_path = (
    "/content/drive/MyDrive/"
    "Fpolis - Distritos Adm/"
    "gvw_distritos_administrativos.gpkg"
)

gdf = gpd.read_file(shp_path)
gdf.to_file(gpkg_path, layer="distritos_administrativos", driver="GPKG")

print("Conversão concluída com sucesso!")
```
